```python
import numpy as np
from scipy import ndimage,misc
import matplotlib.pyplot as plt
```

```python
mapache = misc.face()
type(mapache)
```

```python
mapache.shape
mapache = np.delete(mapache,1,axis=1)
mapache.shape
mapache.append()
```

```python
plt.imshow(mapache)
plt.show()
```

```python
N = 2
arregloCuadros = []
for linea in np.hsplit(mapache,N):
    print(linea.shape)
    arregloCuadros.append(np.split(linea,N))
    print(len(arregloCuadros))
arregloCuadros = np.array(arregloCuadros)
```

```python
arregloCuadros.shape
```

```python
aux = arregloCuadros[0][0].copy()
arregloCuadros[0][0] = arregloCuadros[0][1].copy()
arregloCuadros[0][1] = aux.copy()
```

```python
plt.imshow(arregloCuadros[0][1])
plt.show()
```

```python
lineas = arregloCuadros.reshape(2,768,512,3)
plt.imshow(lineas[1])
plt.show()
```

```python
imagenNueva = np.stack(lineas,axis=1).reshape(768,1024,3)
plt.imshow(imagenNueva)
plt.show()
```
